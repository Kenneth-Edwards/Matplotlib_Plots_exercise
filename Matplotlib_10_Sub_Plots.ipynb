```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data1.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# # use this code to instantiate 2 figures and with 1 plot each
# fig1, ax1 = plt.subplots(figsize=(10, 10))
# fig2, ax2 = plt.subplots(figsize=(10, 10))

# instantiate 1 figure and with 2 plots (side by side)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 5), sharex=True)

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

print(fig)
print(fig1)
print(fig2)
print(ax1)
print(ax2)

plt.show()

#save figs to file
fig1.savefig('plot_10a.png')
fig2.savefig('plot_10b.png')
```
