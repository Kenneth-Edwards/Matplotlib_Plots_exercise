```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data5.txt')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.figure(figsize=(10,8))

# first line plot
plt.plot(ages, dev_salaries, color='#444444',
        linestyle='--', label='All Devs')

# Second line plot
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

'''The parameters y1 and y2 can be scalars, indicating a horizontal boundary at the given y-values. If only y1 is given, y2 defaults to 0.'''
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/fill_between_demo.html
# https://www.geeksforgeeks.org/matplotlib-pyplot-fill_between-in-python/ 

# plt.fill_between(ages, py_salaries, overall_median,
#                  where=(py_salaries > overall_median), # only fills above median salary
#                  interpolate=True, alpha=.25)

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries), # only fills above All dev salary
                 interpolate=True, alpha=.25, label='Above_Avg.')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries), # only fills below All dev salary
                 interpolate=True, color='red', alpha=.25, label='Below_Avg.')



plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.savefig('plot_05a.png') # save plot to file.png

plt.show()
```
