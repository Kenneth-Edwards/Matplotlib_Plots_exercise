```python
''' Pie Charts with Matplotlib'''
import numpy as np
from matplotlib import pyplot as plt
# %matplotlib Qt 
%matplotlib inline

plt.figure(figsize=(10,8))
plt.style.use("fivethirtyeight")

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors = ['#6d904f', '#fc4f30', '#e5ae37', '#008fd5', ]
explode = [0,0,0,0.1,0]


# plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})
plt.pie(slices, labels=labels,
        explode=explode, # explode out slice to highlight
        shadow = True, # give the plot a shadow
        startangle=90, # starting angle of plot
        autopct='%1.1f%%', # add the percentages to the pie slices
        wedgeprops={'edgecolor':'black'}) # add an edge color


plt.title("Top Five Programming Languages Survey")

plt.tight_layout() # helps with padding

plt.savefig('plot_03a.png') # save plot to file.png

plt.show()
```
